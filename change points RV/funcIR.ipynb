```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rhoH(value):
    rhoH=(-np.power(3,2*value)+np.power(2,2*value+2)-7)/(8-np.power(2,2*value+1))
    return rhoH
```

```python
def lam0r(value):
    #lam0r = (1/np.pi)*np.arccos(-value)
    lam0r = (2/np.pi)* np.arctan(np.sqrt((1+value)/(1-value)))+(1/np.pi)*np.log(2/(1+value))*(np.sqrt((1+value)/(1-value)))
    return lam0r
```

```python
def lamH(value):
    lamH=lam0r(rhoH(value))
    return lamH
```

```python
H=np.linspace(0,0.9,20)
lam=lamH(H) 
```

```python
# Detect by ARIMA
# 创建一个新的图形对象和一个子图对象
fig, ax = plt.subplots()

# 绘制一条曲线
x=H
y=lam
ax.plot(x,y,color='#800000', linestyle='-.',linewidth=2)

plt.xlabel('$H$')
plt.ylabel('$\Lambda(H)$')




# 显示图形

plt.savefig('FIG/Lam.png', dpi=300)
# 添加标题

plt.show()
```
